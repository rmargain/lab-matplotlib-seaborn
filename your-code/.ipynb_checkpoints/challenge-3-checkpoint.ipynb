#### Use the appropriate visulalization to show the distribution of the column `Age`

#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
# your code here
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize = (12, 18))
ax1.title.set_text('10 Bin Age Histogram')
sns.histplot(data=titanic, x='Age', bins=10, ax=ax1)
ax2.title.set_text('20 Bin Age Histogram')
sns.histplot(data=titanic, x='Age', bins=20, ax=ax2)
ax3.title.set_text('50 Bin Age Histogram')
sns.histplot(data=titanic, x='Age', bins=50, ax=ax3)
plt.tight_layout()
plt.show()
```

#### Whether the bin size affects your plot? Why?

#your comment here
As we increase the number of bins our individual age ranges reduce and we get closer and closer to the exact age of passengers. Plots do look different given the reason above.

#### Using Seaborn plot the correlation matrix of various features.

*Hint: search how to use `heatmap`*

```python
#your code here
mask = np.triu(np.ones_like(titanic.select_dtypes(['int64', 'float64']).corr(), dtype=bool))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(titanic.select_dtypes(['int64', 'float64']).corr(), mask= mask, cmap=cmap)
```

#### According to the previous plot, what are the most correlated feature? Comment.

#your comment here
Most positively correlated features are: Parch and SibSp, folloewd by Fare and Survived, and Parch and Fare
Most negatively correlated features (which do have the highest correlation) are: Fare and Pclass, and Pclass and Age
